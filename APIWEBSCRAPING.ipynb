```python
import requests
```

```python
#2.Store API URL
```

```python
api_url="https://publicapis.io/_next/data/build/category/development.json?category_id=development"
# api_url="https://jsonplaceholder.typicode.com/posts"
```

```python
#3.Send GET request from url
```

```python
response=requests.get(api_url)
```

```python
print(response.status_code)
```

```python
#4.Convert response into json format
```

```python
data=response.json()
```

```python
print(data)
```

```python
print(data.keys())
```

```python
import pandas as pd
api_list = data['pageProps']['categoryApis']
records=[]
for f in api_list:
    t = f.get('title')
    d = f.get('description')
    c = f.get('category')
    cid = f.get('category_id')
    s = f.get('slug')
    w = f.get('website')
    l = f.get('logolink')
    records.append({
        "Title": t,
        "Description": d,
        "Category": c,
        "Category_id": cid,
        "Slug": s,
        "Website": w,
        "Logolink": l
    })

    print("Title:", t)
    print("Description:", d)
    print("Category:", c)
    print("Category_id:", cid)
    print("Slug:", s)
    print("Website:", w)
    print("Logolink:", l)
    print("-" * 40)
     
df=pd.DataFrame(records)
print(df.head())
# for d in data:
#     u=d['userId']
#     i=d['id']
#     t=d['title']
#     b=d['body']
#     print("UserID:",u)
#     print("ID:",i)
#     print("Title:",t)
#     print("Body:",b)
#     print("-"*100)
        
```

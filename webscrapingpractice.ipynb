```python
api_url="https://edugo-api.aspiria.ai/aspiria/course-finder/?page=1&page_size=16"
response=requests.get(api_url)
```

```python
response.status_code
```

```python
data=response.json()
```

```python
data
```

```python
data.keys()
```

```python
courses = data['data']['university_courses']
for i in courses:
    ucid = i.get("university_course_id")
    uname = i.get("university")
    uid = i.get("university_id")
    country = i.get("country")
    duration = i.get("duration")
    currency = i.get("currency")
    title = i.get("title")
    city = i.get("city")
    tuition = i.get("tuitionFees")
    print("University course id:", ucid)
    print("University Name:", uname)
    print("UniversityID:", uid)
    print("Country:", country)
    print("Duration:", duration)
    print("Currency:", currency)
    print("Title:", title)
    print("City:", city)
    print("Tuition:", tuition)
    print("-" * 40)            
```
